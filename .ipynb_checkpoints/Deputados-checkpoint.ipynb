```python
df_dados = pd.read_csv('gastos-deputados.csv', sep=';')
```

```python
df_dados.head()
```

```python
qt_candidatos_uf = df_dados[['sgUF','ideCadastro']].groupby('sgUF', as_index=False).count()
qt_candidatos_uf
```

```python
qt_candidatos = df_dados[['sgUF','ideCadastro']].drop_duplicates()
qt_candidatos_uf = qt_candidatos.groupby('sgUF', as_index=False).count()
qt_candidatos_uf
```

```python
!pip install altair
```

```python
import altair as alt
```

```python
alt.Chart(qt_candidatos_uf, title='titulo bla bla').mark_bar().encode(
    x=alt.X('sgUF', title='Estado'),
    y=alt.Y('ideCadastro', title='candidatos por estado'),
    color= alt.Color('sgUF')
)
```

```python
qt_candidatos_part = df_dados[['sgPartido','ideCadastro']].drop_duplicates()
qt_candidatos_partido = qt_candidatos_part.groupby('sgPartido', as_index=False).count()
qt_candidatos_partido
```

```python
alt.Chart(qt_candidatos_partido, title='titulo bla bla2').mark_bar().encode(
    x=alt.X('sgPartido', title='Partido'),
    y=alt.Y('ideCadastro', title='candidatos por partido'),
    color= alt.Color('sgPartido')
)
```

```python
uf_valor = df_dados[['sgUF','vlrDocumento']].groupby('sgUF', as_index=False).sum()
#uf_valor_decresc = uf_valor.sort_values('vlrDocumento',ascending=False).head(5)
uf_valor_top5 = uf_valor.nlargest(5,'vlrDocumento')
uf_valor_top5
```

```python
uf_valor_top5_index = uf_valor_top5.reset_index()
uf_valor_top5_index[['sgUF','vlrDocumento']]
```

```python
descricao_parlamentar = df_dados[['txtDescricao','txNomeParlamentar']].groupby('txtDescricao',as_index=False).count()
descricao_parlamentar_top5 = descricao_parlamentar.nlargest(5,'txNomeParlamentar')
descricao_parlamentar_top5
```

```python
descricao_valor = df_dados[['txtDescricao','vlrLiquido','vlrDocumento']].groupby('txtDescricao',as_index=False).sum()
descricao_valor_decresc = descricao_valor.sort_values('vlrDocumento',ascending= False)
descricao_valor_decresc
```

```python
ano_menor_2016 = df_dados['numAno'] < 2016
df_dados[ano_menor_2016]
```

```python
leg_menor_2016 = df_dados['nuLegislatura'] < 2016
df_dados_menor_2016 = df_dados[leg_menor_2016]
qt_candidatos_part_2016 = df_dados_menor_2016[['sgPartido','ideCadastro']].drop_duplicates()
qt_candidatos_partido_2016 = qt_candidatos_part_2016.groupby('sgPartido', as_index=False).count()
qt_candidatos_partido_2016
```

```python
alt.Chart(qt_candidatos_partido_2016, title='titulo bla bla < 2016').mark_bar().encode(
    x=alt.X('sgPartido', title='Partido'),
    y=alt.Y('ideCadastro', title='candidatos por partido'),
    color= alt.Color('sgPartido')
)
```
